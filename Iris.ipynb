```python
#sepallength,sepalwidth

for i in range(20):
  sample1=df.sepalLength.sample(frac=0.04)
  sample2=df.sepalWidth.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)

  if p_value> 0.5:
    print("H0 is accepted")
  else:
    print("H0 is rejected")
```

```python
#petallenght,petallwidth

for i in range(20):
  sample1=df.petalLength.sample(frac=0.04)
  sample2=df.petalWidth.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)

  if p_value> 0.5:
    print("H0 is accepted")
  else:
    print("H0 is rejected")
```

```python
#sepalength and petallength

for i in range(20):
  sample1=df.sepalLength.sample(frac=0.04)
  sample2=df.petalLength.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)

  if p_value> 0.5:
    print("H0 is accepted")
  else:
    print("H0 is rejected")
```

```python
#sepallength and petalwidth
for i in range(20):
  sample1=df.sepalLength.sample(frac=0.04)
  sample2=df.petalWidth.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)

  if p_value> 0.5:
    print("H0 is accepted")
  else:
    print("H0 is rejected")
```

```python
#sepalwidth and petallength
for i in range(20):
  sample1=df.sepalWidth.sample(frac=0.04)
  sample2=df.petalLength.sample(frac=0.04)
  t_test,p_value=stats.ttest_ind(sample1,sample2)

  if p_value> 0.5:
    print("H0 is accepted")
  else:
    print("H0 is rejected")
```
